### Para analisar os dados vou utilizar a biblioteca **pandas** e para gerar gráficos a **matplotlib**

```python
import pandas as pd
import matplotlib

#Importação dataset
uriCipa = pd.read_csv("https://raw.githubusercontent.com/EdsonMSouza/08_T22_QUI_N/main/CENIPA/ocorrencias_cenipa.csv",sep =';',encoding='ISO-8859-1')
uriCipa.head()
```

### Utilizei o **filter** para apresentar colunas específicas e o **columns** para alterar o nome das colunas. 

```python
colunasSelecionadas = ['tipo','descricao','ano','aeronave_liberada','aeronave_ano_fabricacao']
microdadosCipa = uriCipa.filter(items = colunasSelecionadas)
microdadosCipa.columns = ['Evento','Descricao','Ano','Liberado','Fabricacao da Aeronave']
microdadosCipa.head(21)
```

```python
print('Ano de fabricação das aeronaves:')
microdadosCipa['Fabricacao da Aeronave'].unique()
```

```python
microdadosCipa.plot(Kind='hist')
```
